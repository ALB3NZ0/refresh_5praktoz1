Далее была проведена выборка данных по строкам и столбцам с помощью loc.

selected_data = starwars.loc[0:5, ["name", "height", "mass", "films"]] использует метод loc для создания подвыборки данных из исходного набора данных starwars.

Метод loc:

loc - это метод для индексации и выборки данных в DataFrame на основе меток (имен строк и столбцов).
Использование loc для выбора строк и столбцов:

starwars.loc[0:5, ["name", "height", "mass", "films"]] означает следующее:
0:5 - это диапазон индексов строк от 0 до 5 (включительно). Таким образом, выбираются строки с индексами 0, 1, 2, 3, 4 и 5.
["name", "height", "mass", "films"] - это список названий столбцов, которые мы хотим выбрать из каждой выбранной строки.
Результат выборки:

Результат выборки сохраняется в переменной selected_data.
Выбранные столбцы включают "name" (имя персонажа), "height" (рост), "mass" (масса) и "films" (фильмы, в которых участвовал персонаж).
Вывод selected_data:

Вывод selected_data представляет собой небольшую подвыборку данных, состоящую из первых шести строк (индексы 0-5) и выбранных столбцов.
Каждая строка содержит информацию о имени персонажа, их росте, массе и фильмах, в которых они участвовали.
